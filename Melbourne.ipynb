```python
import pandas as pd
import re
```

```python
df = pd.read_csv('Melbourne/House.csv')
df
```

```python
filter1 = df.loc[(df['Rooms'] == 3) & (df['Type'] == 'h') & (df['Price'] <= 250000)]
filter1
```

```python
filter2 = df.loc[(df['Address'].str.contains('Turner.St|Turner Rd', flags=re.I))]
filter2
```

```python
filter3 = df.loc[df['Address'].str.contains('^59') & (df['Price'] <= 500000)]
filter3
```

```python
df.loc[df['SellerG'] == 'Nelson', 'SellerG'] = 'Ronaldo'
df.loc[df['SellerG'] == 'Ronaldo', ['Method']] = 'Pedding'
filter4 = df.loc[df['SellerG'].str.contains('Ronaldo')]
filter4
```

```python
df.groupby(['SellerG']).count().sort_values('Rooms', ascending = False).head(10)
df.groupby(['SellerG']).mean().sort_values('Rooms', ascending = False).head(10)
df.groupby(['SellerG']).sum().sort_values('Rooms', ascending = False).head(10)
```
